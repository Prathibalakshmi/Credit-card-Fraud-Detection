### Input Split

```python
X=df.drop(columns=['Class'],axis=1)
y=df['Class']
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaler=sc.fit_transform(X)
```

```python
x_scaler[-1]
```

### Model Training

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=42,stratify=y)
```

```python
one_hot_encoded1=pd.get_dummies(y_train)
one_hot_encoded1
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,f1_score
model=LogisticRegression()
#training
model.fit(X_train,y_train)
#testing
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
print("F1 Score ",f1_score(y_test,y_pred))
```

### Inference
*There are two class  variables 0 and 1
*Support are the samples that we are taking for testing
*Accuracy is 100% for Logistic Regression because of standard scaler, we cannot go for it as majority of them are non fraudulent samples.
*When a class is imbalanced; 
*Precition-- considers all positive samples that are considered either correctly or in correctly(metric that measures the accuracy of positive prediction)
*Recall--- dependent on positive samples and independent of negative sample(measures the completeness of positive predictions)
*F1 score is the combination of precition and recall

### RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,f1_score
model=RandomForestClassifier()
#training
model.fit(X_train,y_train)
#testing
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
print("F1 Score ",f1_score(y_test,y_pred))
```

### Class Imbalancement

```python
y_train
```

```python
# Hint--> use combination of over Sampling and Under sampling
#balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample=SMOTE()
x_smote,y_smote=over_sample.fit_resample(X_train,y_train)#instead of going to the whole data we take trained data to get better result
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,f1_score
model=LogisticRegression()
#training
model.fit(x_smote,y_smote)
#testing
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
print("F1 Score ",f1_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,f1_score
model=RandomForestClassifier(n_jobs=-1)
#training
model.fit(x_smote,y_smote)
#testing
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
print("F1 Score ",f1_score(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,f1_score
model=XGBClassifier(n_jobs=-1)
#training
model.fit(x_smote,y_smote)
#testing
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
print("F1 Score ",f1_score(y_test,y_pred))
```
